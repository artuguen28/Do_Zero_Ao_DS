#### Respondendo as perguntas de negócio

Data do imóvel mais antigo

```python
data['date'] = pd.to_datetime(data['date'])
data.sort_values('date', ascending=True)
```

Determinar o maior numero de andares e contar quantos temos por andar

```python
data['floors'].unique()

print(data.loc[data['floors'] == 3.5].shape)
```

Criando classificação

```python
data['level'] = 'standard'

data.loc[data['price'] > 540000, 'level'] = 'high_level'
data.loc[data['price'] < 540000, 'level'] = 'low_level'

data.head()
```

Relatório ordenado pelo preço

```python
report = data[['id', 'date', 'price', 'bedrooms', 'sqft_lot', 'level']].sort_values('price', ascending=False)
report.to_csv('datasets/report_aula02.csv', index=False)
```
